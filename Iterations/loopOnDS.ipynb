**Loop on dictionary**

```python
student = {"name": "ali", "age": 19, "year": 1}
for key in student:
    print(key)
```

```python
student = {"name": "ali", "age": 19, "year": 1}
for key in student:
    print(student[key])
```

```python
student = {"name": "ali", "age": 19, "year": 1}
for key in student:
    print(key,":",student[key])
```

```python
students = [
    {"name": "ali", "age": 19, "year": 1},
    {"name": "azeem", "age": 20, "year": 2}
]
for student in students:
    print(f"Name:  {student.get('name','')} | Year: {student.get('year','')}")
```
