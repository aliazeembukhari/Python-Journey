```python
df = pd.read_csv("dataFile-Copy1.csv",
                dtype={
                    "Category":'category',
                    'Color':'category'
                })
print(df.isna().sum())
```

```python
#Fill missing values with the help of mean 
df['Price'] = pd.to_numeric(df['Price'])
df['Price'] = df['Price'].fillna(df['Price'].mode()[0])
```

```python
#Fill missing value with the help of median
df['Stock'] = df['Stock'].astype(str).str.strip().replace('', pd.NA)
df['Stock'] = pd.to_numeric(df['Stock'])
df['Stock'] = df['Stock'].fillna(df['Stock'].median())
```

```python
#Fill the missing value with the help of mode
df['Category'] = df['Category'].fillna(df["Category"].mode()[0])
```
