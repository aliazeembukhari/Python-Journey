```python
import pandas as pd
```

```python
#creating Series
s = pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'])
print(s)
```

```python
print('Values:',s.values)
print('Index:',s.index)
print("Data type:",s.dtype)
```
