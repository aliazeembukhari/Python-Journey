```python
import pandas as pd
```

```python
data = {
    "Name":["Ali","Azeem","Bukhari"],
    "Age":[18,19,20],
    "City":["Multan","Bwp","Vehari"],
    "Salary":[100000,120000,123000]
}
df = pd.DataFrame(data)
print("Orignal Data:\n",df)
```

```python
subset_loc = df.loc[1:2, ['Name', 'Salary']] #at first we are selecting rows and then columns by giving name 
print("Subset with .loc\n", subset_loc)
```
