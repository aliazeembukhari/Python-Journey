```python
import pandas as pd
```

```python
data = {
    "Name":["Ali","Azeem","Bukhari"],
    "Age":[18,19,20],
    "City":["Multan","Bwp","Vehari"],
    "Salary":[100000,120000,123000]
}
df = pd.DataFrame(data)
print("Orignal Data:\n",df)
```

```python
subset_loc = df.loc[1:2, ['Name', 'Salary']] #at first we are selecting rows and then columns by giving name 
print("Subset with .loc\n", subset_loc)
```

```python
subset_iloc = df.iloc[0:4, 0:4] #here we will give both rows no and columns no in index form
print("Subset with .iloc\n",subset_iloc)
```

```python
#boolean filtrint
highSalary = df[df['Salary']>120000-1]
print("High Salary Filter:\n",highSalary)
```

```python
df.loc[df['Age']>19,"Salary"] =99999
print("Low age Salary change: \n",df.loc[df['Age']>19,"Salary"])
```
