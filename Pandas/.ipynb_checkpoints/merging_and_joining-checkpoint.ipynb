```python
import pandas as pd
```

```python
customers = pd.DataFrame(
    {'Name':["Ali",'Ahmad','Talha'],
    'CustomerID':[1,2,3]}
)
orders = pd.DataFrame({
    'OrderID':[101,102,103,104],
    'CustomerID':[1,2,2,3],
    'Amount':[110,150,200,250]
})
```

```python
#Inner merging
df_inner = pd.merge(customers,orders,on="CustomerID", how='inner')
print('Inner Merging:\n',df_inner)
```

```python
#left merging
df_left = pd.merge(customers,orders,on="CustomerID",how= 'left')
print("Left Merging:\n",df_left)
```

```python
#Concatenate Rows
df_concat = pd.concat([customers,customers])
print('Concatenated Rows:\n',df_concat)
```
