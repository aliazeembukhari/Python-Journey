```python
try:
    with open("unknownfile","r") as file:
        print(file.read())
except FileNotFoundError:
    print("File not found")
```

```python
filename = "NewFile"
name = "Ali Azeem"
age = 19
```

```python
try:
    with open(filename,"r") as file:
        print(file.read())
except FileNotFoundError:
    with open(filename,"w") as fout:
        fout.write(f"Hi I am {name} \nI am {age} years old")
    with open(filename,"r") as fin:
        print(fin.read())
```
