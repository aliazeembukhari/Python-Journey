```python
import pandas as pd
import numpy as np
df = pd.read_csv("telecom_customer_churn_feature_engineering.csv")
```

```python
#Date part exraction
df['signup_date'] = pd.to_datetime(df['signup_date'])

df['signup_year'] = df['signup_date'].dt.year
df['signup_month'] = df['signup_date'].dt.month
df['signup_day'] = df['signup_date'].dt.dayofweek
df[['signup_date','signup_year','signup_month','signup_day']].head()
```

```python
#Ratio feature
df['bill_to_income_ratio'] = (
    df['monthly_bill'] / df['monthly_income']
)

df[['monthly_bill','monthly_income','bill_to_income_ratio']].head()
```

```python
#interaction feature
df['support_bill_interaction'] = (
    df['support_tickets'] * df['monthly_bill']
)
df[['support_tickets','monthly_bill','support_bill_interaction']].head()
```

```python
df['feedback_length'] = df['customer_feedback'].str.len()

df[['customer_feedback','feedback_length']].head()
```

```python
#keyword flag feature
df['complaint_flag'] = df['customer_feedback'].str.contains(
    'slow|poor|expensive|unrealiable|drops',
    case = False
).astype(int)

df[['customer_feedback','complaint_flag']].head()
```
