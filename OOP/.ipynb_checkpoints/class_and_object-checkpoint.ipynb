```python
class Bus:
    def drive(self):
        print("The bus is on the road")
        
```

```python
bus = Bus()
bus.drive()
```

```python
class Car:
    color = "Blue"
    def start(self):
        print(f"The {self.color} is started")
```

```python
myCar = Car()
myCar.start()
```

```python
class Car1:
    color = "Red"
    def drive(self):
        print(f"The {self.color} is on the Road")
    def setColor(self,new_color):
        self.color = new_color
        
    
```

```python
car = Car1()
car.setColor("Purple")
car.drive()
```
