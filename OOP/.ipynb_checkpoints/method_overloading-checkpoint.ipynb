```python
class Animal:
    def bark(self):
        print("Dog Barking")
    def roar(self):
        print("Lion Roaring")
```

```python
class Dog(Animal):
    def pet(self):
        print("My dog name is Jack")
    def bark(self):
        print("Jack is Barking")
```

```python
dog1 = Dog()
dog1.bark()
```
