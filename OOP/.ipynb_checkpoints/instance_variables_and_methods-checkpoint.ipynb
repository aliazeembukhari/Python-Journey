```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def display(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Grade: {self.grade}")

    def eligibility(self):
        if self.age>4:
            print(self.name, "is eligible for the admission")
        else:
            print(self.name, "is not eligible for the admission")
```

```python
#creating objects
student1 = Student("Ali",5,"PlayGroup")
student2 = Student("Azeem",3,"PlayGroup")
```

```python
#accening attributes
print(student1.name)
print(student2.name)
```

```python
#calling methods
student1.display()
student1.eligibility()
```

```python
student2.display()
student2.eligibility()
```
