```python
class Animal:
    def bark(self):
        print("Dog Barking")
    def roar(self):
        print("Lion Roaring")
```

```python
class Dog(Animal):
    pass
```

```python
dog1 = Dog()
```

```python
dog1.bark()
```

```python
class Dog(Animal):
    def pet(self):
        print("My dog name is Jack")
```

```python
dog2 = Dog()
dog2.pet()
```

```python
dog2.bark()
```
