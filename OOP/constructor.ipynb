```python
class Dog:
    def __init__(self):
        self.name = "Jack"
    def bark(self):
        print(f"{self.name} is barking")
```

```python
dog = Dog()
dog.bark()
```

```python
class Book:
    def __init__(self, brand, title):
        self.brand = brand 
        self.title = title
    
```

```python
book1 = Book("New Books", "Philosophy")
print(book1.brand)
print(book1.title)
```

```python
book1.title = "Novels"
book2 = Book("old Books", "Sonets")
print(book1.title)
print(book2.title)
print(book2.brand)
```
