```python
class Student:
    @property
    def grade(self):
        return self.__grade
    def __init__(self,name,grade):
        self.__grade = grade
    def get_grade(self):
        return self.__grade
    def set_grade(self,grade):
        if 0 <= grade <=100:
            self.__grade = grade
        else:
            print("Invalid Grade")
```
