```python
class Book:
    def __init__(self, title, author):
        self.title = title 
        self.author = author
    def __str__(self):
        return f"This {self.title} is written by {self.author}"
        
```

```python
b = Book("last night", "Floyd")
```

```python
print(b)
```

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs
    def __len__(self):
        return len(self.songs)
    def __add__(self,other):
        return Playlist(self.songs + other.songs)
```

```python
p1 = Playlist(["Song 1","Song 2"])
```

```python
p2 = Playlist(["Song 3"])
```

```python
print(len(p1))
print(len(p2))
print(len(p1+p2))
```

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __str__(self):
        return f"x = {self.x} y = {self.y}"
    def __add__(self,other):
        if not isinstance(other, Point):
            return NotImplemented
        return Point(self.x+other.x, self.y+other.y)
```

```python
p1 = Point(3,7)
```

```python
p2 = Point(7,2)
```

```python
print(p1+p2)
```
