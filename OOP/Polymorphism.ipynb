```python
class Cat:
    def speak(self):
        print("Meow!")

class Dog:
    def speak(self):
        print("Bark")
```

```python
for pet in [Dog(),Cat()]:
    pet.speak()
```

```python
# method over-riding
class Vehicle:
    def move(self):
        print("Vehicle is moving ")

class Car(Vehicle):
    def move(slef):
        print("Car is moving")
        
```

```python
var1 = Car()
var1.move()
```

```python
class Calculator:
    def add(slef, a, b=0, c=0):
        return a+b+c
```

```python
cal = Calculator()
```

```python
print(cal.add(5))
```

```python
print(cal.add(5,3))
```

```python
print(cal.add(3,4,6))
```
