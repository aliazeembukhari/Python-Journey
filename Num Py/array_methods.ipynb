```python
import numpy as np
```

```python
arr = np.array([[2,5,12],
                [23,9,18],
                [11,13,17]])
```

```python
print(arr[1,0])
print(arr[2,2])
```

```python
print(arr[-1,-2])
print(arr[-2,-2])
```

```python
#if we want to print a specific portion 
print(arr[0:1,0:2]) #here 0 will included but 1 will not similarly 0 will included and 2 will not 
print(arr[0:2,0:2])
```

```python
print(arr[::2,::2]) #means it will skip two rows and 2 coloumns every time it is printing
```

```python
slice_arr = arr[0, :2]
slice_arr[0] = 99
```

```python
print(arr[0]) #orignal will change 
```
