```python
import numpy as np
```

```python
rng = np.random.default_rng()
flips = rng.integers(0,2, size=100000) # 0 for tail and 1 for heads
prob_head=np.mean(flips)
print("Probalilty of the heads: ",prob_head)
#here the value keep changing 
```

```python
#if we want the same value every time we will add seed key-word
rng = np.random.default_rng(seed=40) #the value assigned to seed here is just random 
flips = rng.integers(0,2, size=100000) # 0 for tail and 1 for heads
prob_head=np.mean(flips)
print("Probalilty of the heads: ",prob_head)
```
