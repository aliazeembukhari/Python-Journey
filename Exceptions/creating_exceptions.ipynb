```python
class InvalidAgeError(Exception):
    pass
```

```python
age = int(input("Enter your age: "))
if age<0:
    raise InvalidAgeError(f"Age cannot be negative: {age}")
```

```python
try:
    age = int(input("Enter your age: "))
    if age<0:
        raise InvalidAgeError(f"Age cannot be negative: {age}")
except InvalidAgeError as e:
    print(e)
```
