```python
try:
    num = int(input("Enter a number: "))
    print(10/num)
except ZeroDivisionError:
    print("You cannot divide it by zero")
except ValueError:
    print("You cannot divide it by a character")
except Exception as e:
    print(f"An unexpected error {e} is occuring")
```
