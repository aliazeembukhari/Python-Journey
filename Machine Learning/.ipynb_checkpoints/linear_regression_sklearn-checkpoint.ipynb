```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1],[3],[4],[5],[7]])
Y = np.array([3,5,7,9,11])
```

```python
# Scatter Plot before Regression
plt.figure(figsize=(6,4))
plt.scatter(X,Y, color='blue')
plt.title("Scatter plot before Linear Regression: ")
plt.xlabel('X values')
plt.ylabel('Y values')
# plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
model.fit(X,Y)
y_pred = model.predict(X)

print('Predicted Values: ',y_pred)
```

```python
# Scatter + Best fit line
plt.figure(figsize=(6,4))
plt.scatter(X,Y, color='blue', label='Orignal Data')
plt.plot(X,y_pred, color='red', label='Best Fit Line')
plt.title("Scatter plot with Regression fit line: ")
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
# plt.grid(True)
plt.show()
```

```python
m = model.coef_
b = model.intercept_
```

```python
print('Slope m: ',m)
print('Y intercept b: ',b)
```
