```python
import numpy as np
```

```python
learning_rate = 0.01
epoch = 5 # no of iteration
X = np.array([1,2,3,4,5]) #input data
Y = np.array([3,4,2,5,6]) #target data
```

```python
m,b = 0.0, 0.0 #initial perimeters 
n = len(X)
```

```python
for i in range(epoch):
    Y_pred = m*X + b
    cost = ((Y_pred - Y)**2).mean()
    derivative_m = (-2/n) * sum(X*(Y-Y_pred))
    derivative_b = (-2/n) * sum(Y-Y_pred)
    m -= learning_rate * derivative_m #derivative w.r.t m
    b -= learning_rate * derivative_b #derivative w.r.t b
    print(f'Epoch {i+1}/{epoch}, Cost: {cost}, m: {m}, b: {b}')
```
