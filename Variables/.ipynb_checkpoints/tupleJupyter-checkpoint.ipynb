**Tuple is an ordered but unchangeable collection
ordered means we cannot modify after creation**

```python
sections = ("red", "blue", "green")
print(sections)
```

**Accessing Elements**

```python
print(sections[1])
```

**Tuple Methods**

```python
numbers = (1,1,3,4,5,6,1)
print(numbers.count(1)) #it will tell how many times 1 is repeated 
```

```python
print(numbers.index(6)) # it will tell us the index of required element |
```
