**Lists are muteable Means we can change the elements of the list**

```python
fruits = ["apple", "banana", "mango", "apple","pineApple"]
print(fruits)
```

**Accessing Elements**

```python
print(fruits[0])
print(fruits[2])
print("Printing the last element if the list is too long "+fruits[-1])
```

**Important List Methods**

```python
fruits.append("orange") #add new element 
print(fruits)
```

```python
fruits.insert(3,"Guava") #adding new element at certain index
print(fruits)
```

```python
fruits.remove("apple") #remove element of this name 
print(fruits)
```

```python
#if we want to remove the 2nd apple we have to give remove comand twice
```

```python
fruits.pop() # it removes the last element of the list 
```

```python
print(fruits)
```

```python
fruits.sort() #sort the list 
print(fruits)
```

```python
for i in fruits: 
    print(i)
```
