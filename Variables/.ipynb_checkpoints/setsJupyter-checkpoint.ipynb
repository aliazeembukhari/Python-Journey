**Sets stores Unique Unordered Values It means there should be no repetition**

```python
mySet = {1,2,2,3,4,5}
print(mySet) # it will ignore the repetition 
```

**Set Methods**

```python
mySet.add(11)
mySet.remove(2)
print(mySet)
```

**Set Operations**

```python
mySet2 = {2,4,6,9}
print(mySet.union(mySet2))
print(mySet.intersection(mySet2))
print(mySet.difference(mySet2))
```
